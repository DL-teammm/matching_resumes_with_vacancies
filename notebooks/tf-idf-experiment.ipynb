```python
import json


def get_lang_info_str(raw_str):
    raw_list = json.loads(raw_str)
    
    if not len(raw_list):
        return ''
    
    lang_str = ''
    
    for raw_dict in raw_list:
        lang_str += f'{raw_dict["code_language"]} '
        if 'level' in raw_dict:
            lang_str += f'{raw_dict["level"]} '
        
    return lang_str.strip()
    

filtered_vacancies_df['language'] = vacancies_df['languageKnowledge'].apply(lambda x: get_lang_info_str(x))
filtered_vacancies_df.head()
```

```python
def get_hard_skills_info_str(raw_str):
    raw_list = json.loads(raw_str)
    
    if not len(raw_list):
        return ''
    
    hard_skills_str = ''
    
    for raw_dict in raw_list:
        hard_skills_str += f'{raw_dict["hard_skill_name"]} '
        
    return hard_skills_str.strip()

filtered_vacancies_df['hard_skills'] = vacancies_df['hardSkills'].apply(lambda x: get_hard_skills_info_str(x))
filtered_vacancies_df.head()
```

```python
def get_soft_skills_info_str(raw_str):
    raw_list = json.loads(raw_str)
    
    if not len(raw_list):
        return ''
    
    soft_skills_str = ''
    
    for raw_dict in raw_list:
        soft_skills_str += f'{raw_dict["soft_skill_name"]} '
        
    return soft_skills_str.strip()

filtered_vacancies_df['soft_skills'] = vacancies_df['softSkills'].apply(lambda x: get_soft_skills_info_str(x))
filtered_vacancies_df.head()
```

# Preprocess data

```python
import spacy

nlp = spacy.load("ru_core_news_sm")
```

```python
import re

def preprocess_text(raw_text):
    if not isinstance(raw_text, str):
        return ''
    
    tokens = [
        token.lemma_.lower() for token in nlp(raw_text) if
        not token.is_stop
        and not token.is_punct
        and not token.is_digit
        and not token.like_email
        and not token.like_num
        and not token.is_space
    ]
    tokens = [re.sub("[^а-яА-Яa-zA-Z]", ' ', item) for item in tokens]
    tokens = [re.sub(r"ё", "е", token) for token in tokens]
    
    return ' '.join(tokens)
```

```python
preprocessed_vacancies_df = pd.DataFrame()

preprocessed_vacancies_df['title'] = filtered_vacancies_df['vacancy_name'].apply(
    lambda x: preprocess_text(x)
)
preprocessed_vacancies_df['education'] = filtered_vacancies_df['education'].apply(
    lambda x: preprocess_text(x)
)
preprocessed_vacancies_df['location'] = filtered_vacancies_df['regionName'].apply(
    lambda x: preprocess_text(x)
)
preprocessed_vacancies_df['salary_min'] = filtered_vacancies_df['salary_min']
preprocessed_vacancies_df['salary_max'] = filtered_vacancies_df['salary_max']
preprocessed_vacancies_df['language'] = filtered_vacancies_df['language'].apply(
    lambda x: preprocess_text(x)
)
preprocessed_vacancies_df['skills'] = filtered_vacancies_df['skills'].apply(
    lambda x: preprocess_text(x)
)

preprocessed_vacancies_df
```

```python
preprocessed_resumes_df = pd.DataFrame()

preprocessed_resumes_df['title'] = filtered_resumes_df['title'].apply(
    lambda x: preprocess_text(x)
)
preprocessed_resumes_df['education'] = filtered_resumes_df['education_level'].apply(
    lambda x: preprocess_text(x)
)
preprocessed_resumes_df['location'] = filtered_resumes_df['city'].apply(
    lambda x: preprocess_text(x)
)
preprocessed_resumes_df['desired_salary'] = filtered_resumes_df['desired_wage']
preprocessed_resumes_df['language'] = filtered_resumes_df['language'].apply(
    lambda x: preprocess_text(x)
)
preprocessed_resumes_df['skills'] = filtered_resumes_df['skills'].apply(
    lambda x: preprocess_text(x)
)

preprocessed_resumes_df.head()
```

# Compute embeddings

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline

text_column_names = ['title', 'education', 'location', 'language', 'skills']

vectors_for_resumes = {}

for col in text_column_names:
    curr_pipe = Pipeline(
        steps=[
            ('counter', CountVectorizer()),
            ('tfidf', TfidfTransformer())
        ]
    ).fit(
#         pd.concat([preprocessed_resumes_df[col], preprocessed_vacancies_df[col]], ignore_index=True)
        preprocessed_vacancies_df[col]
    )
    curr_preds = curr_pipe.transform(preprocessed_resumes_df[col])
    vectors_for_resumes[col] = curr_preds
```

```python
vectors_for_vacancies = {}

for col in text_column_names:
    curr_pipe = Pipeline(
        steps=[
            ('counter', CountVectorizer()),
            ('tfidf', TfidfTransformer())
        ]
    ).fit(
#         pd.concat([preprocessed_resumes_df[col], preprocessed_vacancies_df[col]], ignore_index=True)
        preprocessed_vacancies_df[col]
    )
    curr_preds = curr_pipe.transform(preprocessed_vacancies_df[col])
    vectors_for_vacancies[col] = curr_preds
```

```python
import numpy as np

def compute_final_embeddings(vectors_to_text_cols):
    embeddings = []
    
    count = len(vectors_to_text_cols[list(vectors_to_text_cols.keys())[0]].toarray())
    
    for i in range(count):
        vectors = [preds[i].toarray()[0] for col, preds in vectors_to_text_cols.items()]
        embeddings.append(
            np.concatenate(tuple(vectors))
        )
        
    return embeddings


resume_embs = compute_final_embeddings(vectors_for_resumes)
vacancy_embs = compute_final_embeddings(vectors_for_vacancies)
```

# Cluster result vacancies vectors

## K-Means

```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans


ks = range(1, 150)
ssr = []

for k in ks:
    model = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=48)
    model.fit(
#         np.concatenate(
#             (np.array(resume_embs), np.array(vacancy_embs)),
#         )
        np.array(vacancy_embs),
    )
    ssr.append(model.inertia_)
```

```python
ks = range(1, 150)

plt.subplots(figsize=(45, 10))
plt.plot(ks, ssr, '-o')
plt.xlabel('k')
plt.ylabel('SSE')
plt.xticks(ks)
plt.title('Использование метода локтя для оптимального k')
plt.grid()

plt.savefig('k_sse_tf_idf.png')
```

```python
from sklearn.metrics import davies_bouldin_score

ks = range(2, 150)
db_scores = []
for k in ks:
    model = KMeans(n_clusters=k, init='k-means++', n_init=10, random_state=48)
    clust_labels = model.fit_predict(np.array(vacancy_embs))
    db_score = davies_bouldin_score(np.array(vacancy_embs), clust_labels)
    db_scores.append(db_score)
```

```python
plt.subplots(figsize=(45, 10))
plt.plot(ks, db_scores, '-o')
plt.xlabel('k')
plt.ylabel('Davies-bouldin index')
plt.xticks(ks)
plt.title('Использование индекса Дависа-Булдина для оптимального k')
plt.grid()

plt.savefig('k_db_index_tf_idf.png')
```

```python
import random

# show examples in cluster

def show_cluster_examples(cluster_id: int, max_examples: int = 20):
    print(f'Cluster: {cluster_id}')
    cluster_elems_ids = list(np.argwhere(clust_labels == cluster_id).reshape(-1))
    examples_count = min(max_examples, len(cluster_elems_ids))
    examples = random.sample(cluster_elems_ids, examples_count)

    for i, example_id in enumerate(examples):
#         type_label = 'resume' if example_id < len(resume_embs) else 'vacancy'
        type_label = 'vacancy'
        print(f'example_id: {example_id}, type: {type_label}')

#         title = preprocessed_resumes_df['title'].iloc[example_id] if type_label == 'resume' else preprocessed_vacancies_df['title'].iloc[example_id - len(resume_embs)]
        title = preprocessed_vacancies_df['title'].iloc[example_id]
        print(f'job title: {title}\n')

show_cluster_examples(0)
```
