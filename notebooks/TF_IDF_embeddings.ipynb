# Raw data

```python
RESUMES_DATA_PATH = r'''/content/dst-3.0_16_1_hh_database.csv'''
VACANCY_DATA_PATH = 'vacancies.csv'
```

```python
import pandas as pd

resume_df = pd.read_csv(RESUMES_DATA_PATH, sep=';')
resume_df.head(20)
```

# Preprocess data

```python
def convert_resume_row_to_text(row):
    text = ''

    for col, val in row.items():
        if col == 'Обновление резюме':
            continue

        text += f'{col} {val} '

    return text.strip()

convert_resume_row_to_text(resume_df.iloc[0])
```

```python
def convert_vacancy_row_to_text(row):
    text = ''

    for col, val in row.items():
        if col in ['Emoji', 'Link', 'Hashtags']:
            continue

        text += f'{col} {val} '

    return text.strip()

convert_vacancy_row_to_text(vacancy_df.iloc[0])
```

```python
nlp = spacy.load("ru_core_news_sm")
stopwords = nlp.Defaults.stop_words
print(f'Spacy russian stopwords size: {len(stopwords)}', end='\n\n')
' '.join(stopwords)
```

```python
resume_df_crop = resume_df[:20]

resume_df_crop['text'] = resume_df_crop.apply(
    lambda x: convert_resume_row_to_text(x), axis=1
)

resume_df_crop.head()
```

```python
resume_df_crop['cleaned_text'] = resume_df_crop['text'].apply(
    lambda x: ' '.join(
        token.lemma_.lower() for token in nlp(x) if
        not token.is_stop
        and not token.is_punct
        and not token.is_digit
        and not token.like_email
        and not token.like_num
        and not token.is_space
    )
)

resume_df_crop.head()
```

```python
vacancy_df_crop = vacancy_df[:20]

vacancy_df_crop['text'] = vacancy_df_crop.apply(
    lambda x: convert_vacancy_row_to_text(x), axis=1
)

vacancy_df_crop.head()
```

```python
vacancy_df_crop['cleaned_text'] = vacancy_df_crop['text'].apply(
    lambda x: ' '.join(
        token.lemma_.lower() for token in nlp(x) if
        not token.is_stop
        and not token.is_punct
        and not token.is_digit
        and not token.like_email
        and not token.like_num
        and not token.is_space
    )
)

vacancy_df_crop.head()
```

# TF-IDF vectors

```python
resume_train = resume_df_crop['cleaned_text']
vacancy_train = vacancy_df_crop['cleaned_text']

resume_train.head()
```

```python
vacancy_train.head()
```

```python
X_train = pd.concat([resume_train, vacancy_train], ignore_index=True)

X_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer,  TfidfVectorizer
from sklearn.pipeline import Pipeline

pipe = Pipeline(
    steps=[
        ('counter', CountVectorizer()),
        ('tfidf', TfidfTransformer())
    ]
).fit(X_train)
```

```python
preds = pipe.transform(X_train)

preds
```

# Nearest vectors

## Cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity

vacancy_vectors = pipe.transform(vacancy_train)
resume_vectors = pipe.transform(resume_train)

cosine_similarities = cosine_similarity(vacancy_vectors, resume_vectors)
cosine_similarities
```

```python
import numpy as np

vacancy_to_nearest_resume = np.argmax(cosine_similarities, axis=1)

for i in range(len(vacancy_train)):
  print(f'VACANCY text: {vacancy_train.iloc[i]}')

  nearest_resume_idx = vacancy_to_nearest_resume[i].item()
  print(f'nearest resume has this text: {resume_train[nearest_resume_idx]}')
```

## Euclidean distance

```python
from sklearn.metrics.pairwise import euclidean_distances

euclidean_dists = euclidean_distances(vacancy_vectors, Y=resume_vectors)
euclidean_dists
```

```python
vacancy_to_nearest_resume_euc = np.argmin(euclidean_dists, axis=1)

for i in range(len(vacancy_train)):
  print(f'VACANCY text: {vacancy_train.iloc[i]}')

  nearest_resume_idx = vacancy_to_nearest_resume_euc[i].item()
  print(f'Euclidean nearest resume has this text: {resume_train[nearest_resume_idx]}')
```

## Manhattan distance

```python
from sklearn.metrics.pairwise import manhattan_distances

manhattan_dists = manhattan_distances(vacancy_vectors, Y=resume_vectors)
manhattan_dists
```

```python
vacancy_to_nearest_resume_manh = np.argmin(manhattan_dists, axis=1)

for i in range(len(vacancy_train)):
  print(f'VACANCY text: {vacancy_train.iloc[i]}')

  nearest_resume_idx = vacancy_to_nearest_resume_manh[i].item()
  print(f'Manhattan nearest resume has this text: {resume_train[nearest_resume_idx]}')
```
